```python
%matplotlib inline  

import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_ladder(numSimulations, maxGames, winRateSub5, winRateAt5):

    gameArray = []
    
    for s in range(0,numSimulations):
        starArray = []
        gameIndex = 0
        numStars = 0
        winStreak = 0
        while(gameIndex < maxGames and numStars < 95):
            if(numStars < 70):
                winRate = winRateSub5
            else:
                winRate = winRateAt5

            win = np.random.binomial(1, winRate)
            if(win):
                numStars += 1
                winStreak += 1
                if(winStreak >= 3 and numStars < 70):
                    numStars += 1
            elif(numStars > 10):
                numStars -= 1
                winStreak = 0
            else:
                winStreak = 0    
            gameIndex += 1
            starArray.append(numStars)
        gameArray.append(gameIndex)

    return gameIndex, gameArray, starArray  
```

```python
numSimulations = 1000
maxGames = 10000
winRate = 0.60
winRateSub5 = 1
winRateAt5 = 0.55
```

```python
numGamesMu = []
numGamesSigma = []
thisGameIndex = []
for winRateAt5 in np.arange(0.55, 1, 0.05):
    thisGameIndex, thisGameArray, thisStarArray = simulate_ladder(numSimulations, maxGames, winRateSub5, winRateAt5)
    numGamesMu.append(np.mean(thisGameArray))
    numGamesSigma.append(np.std(thisGameArray))

# Plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)  

plt.yticks(fontsize=14)  
plt.xticks(fontsize=14) 

plt.xlabel('Win rate above rank 5', fontsize=18)
plt.ylabel('Games to legend', fontsize=18)

plt.xlim(0.5, 1)
plt.ylim(0, 450)

plt.errorbar(np.arange(0.55, 1, 0.05), numGamesMu, yerr=numGamesSigma, fmt='o-', 
             color='#1F77B4', linewidth=2, markersize=10, markeredgecolor='w')    

plt.tick_params(axis='both', which='both', bottom='on', top='off', labelbottom='on', 
                left='on', right='off', labelleft='on', direction='out') 

plt.savefig('myimage.png', format='svg', dpi=100)
```

```python
def nanmean(data, **args):
    return np.ma.filled(np.ma.masked_array(data,np.isnan(data)).mean(**args), fill_value=np.nan)
```

```python
import numpy as np

plt.figure(figsize=(12, 9))  
  
ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["bottom"].set_visible(False)  
ax.spines["right"].set_visible(False)  
ax.spines["left"].set_visible(False)  
ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left() 

plt.tick_params(axis="both", which="both", bottom="off", top="off",  
                labelbottom="on", left="off", right="off", labelleft="on")  
 
plt.title('Number of stars vs. games played', fontsize=22)
plt.xlabel('Games played', fontsize=18)
plt.ylabel('Number of Stars', fontsize=18)

maxEnd = 300
plt.hold(True)

plt.ylim(0, 110)  
plt.xlim(0, maxEnd) 

plt.yticks(range(0, 101, 20), fontsize=14)  
plt.xticks(fontsize=14)  

for y in range(0, 101, 20):  
    plt.plot(range(0, maxEnd), [y] * len(range(0, maxEnd)), "--", lw=0.5, color="black", alpha=0.3)  
    
plt.plot([0, maxEnd], [10, 10], lw='2', color='#D62728')
plt.plot([0, maxEnd], [70, 70], lw='2', color='#FFDD71')
plt.plot([0, maxEnd], [95, 95], lw='2', color='#2CA02C')
    
plt.text(maxEnd+3, 10-1, 'Rank 20', fontsize=14, color='#D62728')  
plt.text(maxEnd+3, 70-1, 'Rank 5', fontsize=14, color='#FFDD71')  

# Similate 
numSimulations = 1
maxGames = 10000
winRate = 0.85
winRateSub5 = winRate
winRateAt5 = winRate

numRuns = 1000
a = np.empty((600,numRuns,))
a[:] = 95

for n in range(0,numRuns):
    gameIndex, gameArray, starArray= simulate_ladder(numSimulations, maxGames, winRateSub5, winRateAt5)
    a[0:len(starArray), n] = starArray
    plt.plot(starArray, lw='2', color='k', ls='-', alpha=0.01)

plt.plot(a.mean(axis=1), lw=3, color='#1F77B4')
plt.text(maxEnd+3, 95-1, 'Legend', fontsize=14, color='#2CA02C')  

plt.savefig('starsVGames.png', pad_inches=0.5)
```
