```python
%matplotlib inline  

import numpy as np
import matplotlib.pyplot as plt
```

```python
numSimulations = 100
maxGames = 10000
winRate = 0.60
winRateSub5 = 0.8
winRateAt5 = 0.55
gameArray = []
starArray = []
```

```python
for s in range(0,numSimulations):
    starArray = []
    gameIndex = 0
    numStars = 0
    winStreak = 0
    while gameIndex < maxGames and numStars < 95:
        if(numStars < 70):
            winRate = winRateSub5
        else:
            winRate = winRateAt5
            
        win = np.random.binomial(1, winRate)
        if(win):
            numStars += 1
            winStreak += 1
            if(winStreak >= 3 and numStars < 70):
                numStars += 1
        elif(numStars > 10):
            numStars -= 1
            winStreak = 0
        else:
            winStreak = 0    
        gameIndex += 1
        starArray.append(numStars)
    gameArray.append(gameIndex)
        
print np.mean(gameArray)
print (np.mean(gameArray)*10)/60
    
```

```python
plt.figure(figsize=(12, 9))  
  
ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["bottom"].set_visible(False)  
ax.spines["right"].set_visible(False)  
ax.spines["left"].set_visible(False)  
ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left() 

plt.ylim(0, 110)  
plt.xlim(0, gameIndex) 

plt.yticks(range(0, 101, 20), fontsize=14)  
plt.xticks(fontsize=14)  

for y in range(0, 101, 20):  
    plt.plot(range(0, gameIndex), [y] * len(range(0, gameIndex)), "--", lw=0.5, color="black", alpha=0.3)  

plt.tick_params(axis="both", which="both", bottom="off", top="off",  
                labelbottom="on", left="off", right="off", labelleft="on")  
 
plt.title('Number of stars vs. games played', fontsize=22)
plt.text(gameIndex+10, 10-1, 'Rank 20 (Rank floor)', fontsize=14, color='#D62728')  
plt.text(gameIndex+10, 70-1, 'Rank 5 (No bonus)', fontsize=14, color='#FFDD71')  
plt.text(gameIndex+10, 95-1, 'Legend', fontsize=14, color='#2CA02C')  


plt.hold(True)
plt.plot([0, gameIndex], [10, 10], lw='2', color='#D62728')
plt.plot([0, gameIndex], [70, 70], lw='2', color='#FFDD71')
plt.plot([0, gameIndex], [95, 95], lw='2', color='#2CA02C')
plt.plot(starArray, lw='2', color='#1F77B4')
```
